##  Data Exploration and cleaning

We shall start by running function to have an overview of the dataset to check the parts that needs cleaning

```python
autos.describe(include='all')
```

In the dataset we notice that ***seller*** and ***offer_type*** columns the values appear to the same and the num_pictures columns the data needs to be investigated further

```python
autos["num_pictures"].value_counts()
```

The picture column seems to have zero values for all the rows.We shall drop this  and ***seller*** and ***offer_type*** columns

```python
autos = autos.drop(["num_pictures", "seller", "offer_type"], axis=1)
```

We are also going to clean the ***price*** column and ***odometer*** which is stored as text.

```python
autos["price"] = autos["price"].str.replace('$','').str.replace(',','').astype(float)
autos["price"].head()
```

```python
autos["odometer"] = autos["odometer"].str.replace(',','').str.replace('km','').astype(float)
autos["odometer"].head()
```

```python
#renaming odometer column
autos.rename({'odometer': 'odometer_km'},axis=1,inplace = True)
autos.columns
```

### Exploring the date Columns

There are a number of columns with date information:

***date_crawled***
***ad_created***
***last_seen***
These are a combination of dates that were crawled, and dates with meta-information from the crawler. The non-registration dates are stored as strings.

```python
autos[['date_crawled','ad_created','last_seen']][0:5]
```

```python
(autos["date_crawled"]
        .str[:10]
        .value_counts(normalize=True, dropna=False)
        .sort_index()
        )
```

It looks like the site was crawled daily over roughly a one month period in March and April 2016. The distribution of listings crawled on each day is roughly uniform.

```python
(autos["ad_created"]
        .str[:10]
        .value_counts(normalize=True, dropna=False)
        .sort_index()
        )
```

The crawler recorded the date it last saw any listing, which allows us to determine on what day a listing was removed, presumably because the car was sold.

The last three days contain a disproportionate amount of 'last seen' values. Given that these are 6-10x the values from the previous days, it's unlikely that there was a massive spike in sales, and more likely that these values are to do with the crawling period ending and don't indicate car sales.

```python
(autos["last_seen"]
        .str[:10]
        .value_counts(normalize=True, dropna=False)
        .sort_index()
        )
```

There is a large variety of ad created dates. Most fall within 1-2 months of the listing date, but a few are quite old, with the oldest at around 9 months.

```python
autos['registration_year'].describe()
```

The year that the car was first registered will likely indicate the age of the car. Looking at this column, we note some odd values. The minimum value is 1000, long before cars were invented and the maximum is 9999, many years into the future.

### Incorrect Registration Year Data

Because a car can't be first registered before the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s.

One option is to remove the listings with these values. Let's determine what percentage of our data has invalid values in this column:

```python
autos = autos[autos["registration_year"].between(1900, 2016)]
```

```python
autos["registration_year"].value_counts(normalize=True).head(10)
```

Most of the cars were registered the first 20 years
